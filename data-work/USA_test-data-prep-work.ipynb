```python
# importing pandas and pathlib to import datasets and form dataframe
import pandas as pd
from pathlib import Path
```

```python
# bringing in the dataframe with path function and read_csv
spotify_df = pd.read_csv(Path("../Resources/spotify_data.csv"))

# checking csv data with .head function
spotify_df.head(10)
```

```python
# checking quality of dataset if there are null values
spotify_df.isnull().sum()
```

```python
# print column names
for col in spotify_df:
    print (col)
```

```python
# dropping the Uri from the dataset
spotify_df = spotify_df.drop(columns=['Uri'])

# selecting US data from original dataframe
spotify_us_df = spotify_df[spotify_df['Country'] == "USA"].copy()

spotify_us_df.head(10)
```

```python
# dropping country from us dataset
spotify_us_df = spotify_us_df.drop(columns = ["Country"]).copy()

#visualizing to check
spotify_us_df.head(10)
```

```python
spotify_us_df=spotify_us_df.drop(columns= ['bing_norm_negative', 'bing_norm_neutral', 'bing_norm_positive','Argentina','Australia','Austria','Belgium','Brazil','Canada','Chile','Colombia','Costa Rica','Denmark','Ecuador','Finland','France','Germany','Global','Indonesia','Ireland','Italy','Malaysia','Mexico','Netherlands','New Zealand','Norway','Peru','Philippines','Poland','Portugal','Singapore','Spain','Sweden','Switzerland','Turkey','UK','USA'])
```

```python
spotify_us_df = spotify_us_df[spotify_us_df.isnull().sum()[spotify_us_df.isnull().sum()==0].keys()].copy()
```

We need to convert these columns into ints

‘Artist_followers’, ‘Track_number’, ‘Tracks_in_album’, ‘danceability’, ‘energy’, ‘key’, ‘loudness’ , ‘mode’, ‘speechiness’, ‘acoustics’, ‘instrumentalness’, ‘liveliness’, ‘valence’, ‘tempo’, ‘duration_ms’, ‘time_signature’ = convert

```python
spotify_us_df.head()
```

```python
number_columns = ['Artist_followers', 'Track_number', 'Tracks_in_album', 'danceability', 'energy', 'key', 'loudness' , 'mode', 'speechiness', 'acoustics', 'instrumentalness', 'liveliness', 'valence', 'tempo', 'duration_ms', 'time_signature']

for column in number_columns:
    spotify_us_df[column] = spotify_us_df[column].astype(float)

# checking to see that we have ints now, seems like we do
spotify_us_df.info()
```
