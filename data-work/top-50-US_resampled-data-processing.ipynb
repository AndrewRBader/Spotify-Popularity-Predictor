```python
# Import required libraries 
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.preprocessing import StandardScaler
import hvplot.pandas
```

```python
# grabbing the whole dataframe for processing
spotify_df = pd.read_csv(Path("../Resources/spotify_data.csv"))
```

```python
# looking at the head and shape
display(spotify_df.head())
display(spotify_df.shape)
```

```python
# dropping selected columns
spotify_df = spotify_df.drop(
    columns=['Album/Single', 'Genre', 'Genre_new', 'Explicit', 'Album', 'Release_date', 'Cluster', 
             'Popu_max', 'Popularity', 'Cluster', 'bing_norm_negative', 'bing_norm_neutral', 
             'bing_norm_positive','Argentina','Australia','Austria','Belgium','Brazil','Canada',
             'Chile','Colombia','Costa Rica','Denmark','Ecuador','Finland','France','Germany',
             'Global','Indonesia','Ireland','Italy','Malaysia','Mexico','Netherlands','New Zealand',
             'Norway','Peru','Philippines','Poland','Portugal','Singapore','Spain','Sweden',
             'Switzerland','Turkey','UK','USA'])
spotify_df.head()
```

Extracting US data

```python
# dropping Uri from the dataset
spotify_df = spotify_df.drop(columns=['Uri','Top10_dummy'])

# selecting US data from original dataframe (best sampled data overall)
spotify_us_df = spotify_df[spotify_df['Country'] == "USA"].copy()

# dropping null columns in dataframe - these columns were created by the creators of this data using thier NLP 
spotify_us_df = spotify_us_df[spotify_us_df.isnull().sum()[spotify_us_df.isnull().sum()==0].keys()].copy()

# drop the Country column
spotify_us_df = spotify_us_df.drop(columns=['Country'])

# view the dataframe
spotify_us_df.head()
```

Dropping title and artist in prep for processing and visualization

```python
# drop title and Artist
spotify_us_df = spotify_us_df.drop(columns = ['Title', 'Artist'])
spotify_us_df
```

Some columns need to be converted from objects to floats (as seen in the USA test data prep work file)

```python
number_columns = ['Artist_followers', 'Track_number', 'Tracks_in_album', 'danceability', 'energy', 'key', 'loudness' , 'mode', 'speechiness', 'acoustics', 'instrumentalness', 'liveliness', 'valence', 'tempo', 'duration_ms', 'time_signature']

for column in number_columns:
    spotify_us_df[column] = spotify_us_df[column].astype(float)
```

Next we are counting the number of observations in this feature data to see if there is any sampling bias needed to take into account necessitating resampling

```python
observation_count = spotify_us_df.groupby('Top50_dummy').count()

observation_count
```

```python
observation_count.hvplot.bar(y='Artist_followers', title='Observation Data (top 50)', ylabel='count', xlabel='Top 50 [0 yes, 1 no]')
```

It does seem as though there a slight oversampling issue, roughly 2 no to 1 yes in top 50. Will use random over sampler to mitigate this

```python
# splitting data into features and target for resampling
spotify_features = spotify_us_df.drop(columns = ['Top50_dummy'])
spotify_target = spotify_us_df['Top50_dummy'] 
```

```python
# Import the RandomOverSampler module form imbalanced-learn
from imblearn.over_sampling import RandomOverSampler

# Instantiate the random oversampler model
# # Assign a random_state parameter of 1 to the model
random_oversampler_model = RandomOverSampler(random_state = 1)

# Fit the original training data to the random_oversampler model
spotify_features, spotify_target = random_oversampler_model.fit_resample(spotify_features, spotify_target)
```

Grouping resampled data to check that counts are more equivalent

```python
observation_data_resampled_df = pd.concat([spotify_features, spotify_target], axis=1)

observation_data_resampled_df
```

```python
observation_resampled_count = observation_data_resampled_df.groupby('Top50_dummy').count()

observation_resampled_count
```

```python
observation_resampled_count.hvplot.bar(y='Artist_followers', title='resampled data (Top 50)', ylabel='count', ylim=(0, 5000))
```

Saving to csv in the Resources Folder

```python
observation_data_resampled_df.to_csv("../Resources/top-50_resampled_df.csv")
```
